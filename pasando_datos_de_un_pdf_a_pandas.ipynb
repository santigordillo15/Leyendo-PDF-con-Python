```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#En el siguiente Data encontraremos los resultados historicos de la Champions league, asi pues el resultado 8-0 se ha dado en 2 ocaciones
import tabula
df =tabula.read_pdf("datos, UefaCL (estadisticas historicas).pdf",pages=17 ) #leemos la pagina 17 del pdf
x =df[0]
resultado_partidos = x[["Result","Games","Goals"]].drop(27,axis=0) #seleccionamos columnas que necesitamos y borramos 1 fila
resultado_partidos.set_index(pd.Index([x for x in range(51)]), inplace =True) #cambiamos los indices
resultado_partidos.rename(columns={"Goals":"Goles totales","Games":"Juegos totales"},inplace=True) #renombramos las columnas
```

```python
#creamos una columna = "Goles x partido"
resultado_partidos["Goles x partido"] = resultado_partidos["Goles totales"] / resultado_partidos["Juegos totales"]
round(resultado_partidos, ndigits=2)
```

```python
resultado_partidos.sort_values("Juegos totales",ascending=True,inplace=True) #ordenamos de menor a mayor
x= resultado_partidos.iloc[:-1,0] #eje x
y = resultado_partidos.iloc[:-1,1] #eje y
resultado_partidos["probabilidad"]=resultado_partidos["Juegos totales"]/resultado_partidos.iloc[50,1] # agregamos nueva columna con la probabilida de ocurrencia
resultado_partidos
```

```python
# crear la columna probabilidad acum
lista = list(resultado_partidos["probabilidad"])
lista.remove(1)
lista.reverse()

acum=[]
suma=0
for x in lista:
    suma += x 
    acum.append(suma)

acum.reverse()
acum.append(np.array(acum).sum())
## agregamos colum 'prob acum' al DF
resultado_partidos["prob acum"] = acum
resultado_partidos
```

```python
#probabilidad x goles totales
resultado_partidos.groupby("Goles x partido")[["probabilidad"]].sum()
#CONCLUSION : es más probable que metan 5 goles que el partido quede 0-0
```

```python
#Graficamos
x= resultado_partidos.iloc[:-1,0] #eje x
y = resultado_partidos.iloc[:-1,1] #eje y
plt.figure(figsize=(15, 10))
plt.bar(x,y,animated=True)
plt.xticks(x, rotation = 90,size='large')
plt.xlabel('resultado final')
plt.ylabel('# de partidos con dicho resultado')
plt.show()
```
